```python
condition1 = ABDF['neighbourhood_group'] == 'Manhattan'
condition2 = ABDF['neighbourhood_group'] == 'Brooklyn'
condition3 = ABDF['neighbourhood_group'] == 'Queens'

s1 = ABDF[condition1 | condition2 | condition3]
s2 = s1.groupby(['neighbourhood_group','neighbourhood']).sum()["Revenue"].sort_values(ascending = False).reset_index()

B = s2.loc[s2['neighbourhood_group'] == 'Brooklyn'].head(3)
M = s2.loc[s2['neighbourhood_group'] == 'Manhattan'].head(3)
Q = s2.loc[s2['neighbourhood_group'] == 'Queens'].head(3)

combined = pd.concat([B,M,Q])

combined
```
